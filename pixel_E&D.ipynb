```python
from PIL import Image
import numpy as np
from google.colab import files

def encrypt_image(input_path, output_path, key):
    image = Image.open(input_path)
    image_array = np.array(image)
    encrypted_array = image_array ^ key
    encrypted_image = Image.fromarray(encrypted_array.astype('uint8'))
    encrypted_image.save(output_path)
    print("Encrypted image saved as", output_path)

def decrypt_image(input_path, output_path, key):
    image = Image.open(input_path)
    image_array = np.array(image)
    decrypted_array = image_array ^ key
    decrypted_image = Image.fromarray(decrypted_array.astype('uint8'))
    decrypted_image.save(output_path)
    print("Decrypted image saved as", output_path)
```

```python
# Upload the original image
uploaded = files.upload()  # Upload "original.png"

# Encrypt the image
encrypt_image("sunflower.jpeg", "encrypted.png", 123)

# Decrypt the image
decrypt_image("encrypted.png", "decrypted.png", 123)

# Download results
files.download("encrypted.png")
files.download("decrypted.png")
```
